```python
import pandas as pd

# Charger la base complète
df = pd.read_csv("Speed_Dating_Data.csv", encoding="latin1")

# 1️⃣ Profil des participants
profile_cols = [
    'iid', 'gender', 'age', 'race', 'field', 'career', 'goal', 
    'date', 'income', 'wave'
]
df_profile = df[[c for c in profile_cols if c in df.columns]]

# 2️⃣ Préférences déclarées
prefs_cols = [c for c in df.columns if '1_' in c or '2_' in c]  # souvent nommées comme attr1_1, sinc1_1, etc.
df_prefs = df[prefs_cols]

# 3️⃣ Notes pendant les rencontres
ratings_cols = ['attr', 'sinc', 'intel', 'fun', 'amb', 'shar']
ratings_cols = [c for c in ratings_cols if c in df.columns]
df_ratings = df[ratings_cols]

# 4️⃣ Résultats (matchs, décisions)
results_cols = ['match', 'dec', 'dec_o', 'like', 'prob', 'met']
results_cols = [c for c in results_cols if c in df.columns]
df_results = df[results_cols]

# 5️⃣ Similarités
similarity_cols = ['int_corr', 'samerace']
similarity_cols = [c for c in similarity_cols if c in df.columns]
df_similarity = df[similarity_cols]

# Sauvegarder les sous-parties si besoin
df_profile.to_csv("participants_profile.csv", index=False)
df_prefs.to_csv("preferences.csv", index=False)
df_ratings.to_csv("ratings.csv", index=False)
df_results.to_csv("results.csv", index=False)
df_similarity.to_csv("similarity.csv", index=False)
```
